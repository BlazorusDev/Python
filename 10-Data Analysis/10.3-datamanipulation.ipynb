```python
# Handling Missing Data
df.isnull().sum()
```

```python
# Filling missing values with the mean of the column
df['Values_new']=df['Value'].fillna(df['Value'].mean())
```

```python
df.head()
```

```python
# Renaming columns
df.rename(columns={'Values_new':'Value_filled'}, inplace=True)
```

```python
df.head()
```

```python
# Chnage column data types
df['Value_filled']=df['Value_filled'].astype(int)
```

```python
df.head()
```

```python
df['Value_filled'] = df['Value'].apply(lambda x: x*2)
```

```python
# Data Aggregation
grouped_mean = df.groupby('Product')['Value'].mean()
```

```python
grouped_mean
```

```python
grouped_mean = df.groupby(["Product",'Region'])["Value"].sum()
```

```python
grouped_mean
```

```python
# aggregation with multiple functions
grouped_agg = df.groupby('Product')['Value'].agg(['mean','sum','max','min'])
grouped_agg
```

```python
# Merging And Joining DataFrames
df1 = pd.DataFrame({
    'ID':[1,2,3,4],
    'Name':['Alice','Bob','Charlie','David']
})  
df2 = pd.DataFrame({
    'ID':[3,4,5,6],
    'Age':[23,34,45,56]
})  
merged_df = pd.merge(df1,df2, on='ID', how='inner')
merged_df
```
